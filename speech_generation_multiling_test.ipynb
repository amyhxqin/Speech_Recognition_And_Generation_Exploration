# Multilingual Speech Generation Test

This notebook explores multilingual speech generation for English and for French.

```python
# for python3
import pyttsx3
```

```python
# one time initialization
engine = pyttsx3.init(driverName='nsss')
```

```python
# set properties
# engine.setProperty('rate', 150) # speed percent
# engine.setProperty('volume', 0.9)  # volume between 0 and 1
```

```python
# print all available voices
voices = engine.getProperty('voices')
for voice in voices:
    print("Voice:")
    print(" - ID: %s" % voice.id)
    print(" - Name: %s" % voice.name)
    print(" - Languages: %s" % voice.languages)
    print(" - Gender: %s" % voice.gender)
    print(" - Age: %s" % voice.age)
```

```python
# set languages (depends on OS)
en_voice_id = "com.apple.speech.synthesis.voice.samantha"
# en_voice_id = "com.apple.speech.synthesis.voice.Alex"
fr_voice_id = "com.apple.speech.synthesis.voice.amelie"
```

For English

```python
engine.setProperty('voice', en_voice_id)
engine.say('Hello. I want to place an order.')
engine.runAndWait()
```

For French

```python
engine.setProperty('voice', fr_voice_id)
engine.say('Bonjour. Je veux passer une commande.')
engine.runAndWait()
```
